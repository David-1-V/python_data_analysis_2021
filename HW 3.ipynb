```python
datatime = pd.read_csv('data-20191112T1252-structure-20191112T1247.csv', encoding='utf-8',usecols=[i for i in range(13)], index_col=0)
datatime
```

```python
df_1=datatime.applymap(lambda x: x.split(','))
df_1=datatime.columns = np.arange(1,13)
datatime
```

```python
#Три датафрейма (три категории дат соответственно)
def before_holidays_0(dates): 
    l = [int(d[:-1]) for d in dates if d.endswith('*')]
    return l if len(l) else [None]

def new_holidays_0(dates): 
    l = [int(d[:-1]) for d in dates if d.endswith('+')]
    return l if len(l) else [None]

def weekends_0(dates):
    return [int(d) for d in dates if not d.endswith('*') and not d.endswith('+')]


before_holidays = df_1.applymap(before_holidays_0)
new_holidays = df_1.applymap(new_holidays_0)
weekends = df_1.applymap(weekends_0)

before_holidays
```

```python
before_holidays = list(filter(None, [pd.Timestamp(year, month, day)
                           if day is not None else None
                           for year in before_holidays.index
                           for month in before_holidays.columns
                           for day in before_holidays.loc[year][month]]))

new_holidays = [Holiday(name='Shifted_holiday', year=year, month=month, day=day)
        if day is not None else None
        for year in new_holidays.index
        for month in new_holidays.columns
        for day in new_holidays.loc[year][month]]

weekends = [Holiday(name='Weekend or Holiday', year=year, month=month, day=day)
        if day is not None else None
        for year in weekends.index
        for month in weekends.columns
        for day in weekends.loc[year][month]]

rules = list(filter(None, [*weekends, *new_holidays]))
```

```python
# Создание календаря
class RussianBusinessCalendar(AbstractHolidayCalendar):
    start_date = datetime.datetime(1999, 1, 1)
    end_date = datetime.datetime(2025, 12, 31)
    rules = rules

russian_calendar = CustomBusinessDay(calendar=RussianBusinessCalendar())

business_month_ends = CustomBusinessMonthEnd(holidays=rules)
```

```python
monthes = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь',
           'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
weekdays = ['Пн', 'Вт', 'Ср', 'Чт', 'Пт', 'Сб', 'Вс']
d_types = ['Выходной', 'Праздник', 'Предпраздничный рабочий день',
           'Последний день месяца', 'Последний рабочий день месяца']
cols = [*d_types, *weekdays, *monthes, 'Предновогодний выходной день']

Work_calendar = pd.DataFrame(columns=cols, index=pd.date_range(start='1999-01-01', end='2025-12-31', freq='D'))
Work_calendar = custom_calendar.fillna(0)
```

```python
holiday_dates = [*['1999-01-0{}'.format(i) for i in range(1, 9)], '1999-02-23', '1999-03-08', '1999-05-01','1999-05-09', 
                 '1999-06-12', '1999-11-04']
official_holidays = [pd.date_range(start=d, end='2025-12-31', freq=DateOffset(years=1)) for d in holiday_dates]
official_holidays = list(chain(*official_holidays))
new_year_eves = pd.date_range(start='2000-12-31', end='2025-12-31', freq=DateOffset(years=1))
bus_month_ends = pd.date_range(start='1999-01-01', end='2025-12-31', freq=business_month_ends)
```

```python
Work_calendar.loc[russian_calendar.holidays, ['выходной']] = 1
Work_calendar.loc[official_holidays, ['праздник']] = 1
Work_calendar.loc[official_holidays, ['выходной']] = 1
Work_calendar.loc[new_year_eves, ['Предновогодний выходной день']] = 1
Work_calendar.loc[custom_calendar.index.is_month_end == True, ['последний день месяца']] = 1
Work_calendar.loc[bus_month_ends, ['последний рабочий день месяца']] = 1
Work_calendar.loc[before_holidays, ['предпраздничный рабочий день']] = 1

for i, v in enumerate(m_names):
    Work_calendar.loc[Work_calendar.index.month == i+1, [v]] = 1

for i, v in enumerate(wd_names):
    Work_calendar.loc[Work_calendar.index.dayofweek == i, [v]] = 1
```

```python
data = pd.read_csv('sber_stocks.csv', usecols=['CLOSE', 'TRADEDATE'], 
                 index_col='TRADEDATE', parse_dates=True)
data = data.CLOSE
```

```python
data_1 = data.asfreq(pd.offsets.Day()).fillna(method='ffill')
data_1
```

```python
data_2 = pd.DataFrame(data_1)
data_2['weekday'] = data_2.index.weekday
data_2
```

```python
for i in range(1, 6, 2):
    data_2['Лаг ' + str(i)]=data_1.shift(i, freq=russian_calendar).values
data_2
```

```python
for i in range(1, 6, 2):
    data_2['Cкользящая статистика ' + str(i) + ' mean'] = data_2.groupby('weekday')['CLOSE'].transform(
        lambda x: x.rolling(window=i).mean().shift(1))
    data_2['Cкользящая статистика ' + str(i) + ' median'] = data_2.groupby('weekday')['CLOSE'].transform(
        lambda x: x.rolling(window=i).median().shift(1))
    data_2['Cкользящая статистика ' + str(i) + ' max'] = data_2.groupby('weekday')['CLOSE'].transform(
        lambda x: x.rolling(window=i).max().shift(1))
    data_2['Cкользящая статистика ' + str(i) + ' min'] = data_2.groupby('weekday')['CLOSE'].transform(
        lambda x: x.rolling(window=i).min().shift(1))
    if i == 1:
        continue
    data_2['Cкользящая статистика ' + str(i) + ' std'] = data_2.groupby('weekday')['CLOSE'].transform(
        lambda x: x.rolling(window=i).std().shift(1))
data_2
```
