```python
from graphviz import Digraph

g = Digraph('G')

with g.subgraph(name='cluster_0') as c:
    c.node('D', label='Фотолиз воды')
    c.node('E', label='Восстановление НАДФ+')
    c.node('F', label='Синтез АТФ')
    c.edges([('D', 'E'), ('E', 'F')])
    c.attr(label='Световая фаза')

with g.subgraph(name='cluster_1') as c:
    c.attr(style='filled', color='lightgrey')
    c.node_attr['style'] = 'filled'
    c.node('A', label='Окисление НАДФ*H2')
    c.node('B', label='Распад АТФ')
    c.node('C', label='Цикл Кальвина')
    c.edges([('A', 'B'), ('B', 'C')])
    c.attr(label='Темновая фаза')
    
with g.subgraph(name='cluster_2') as c:
    c.engine ='circo'
    c.node('G', 'Карбоксилирование')
    c.node('H', 'Восстановление')
    c.node('I', 'Регенерация')
    c.edges([('C', 'G'), ('G', 'H'), ('H', 'I'), ('I', 'C')])
    
    
    
g.edge('start', 'D')
g.edge('start', 'A')

g.node('start', shape='diamond', label='Фотосинтез')




g
```
